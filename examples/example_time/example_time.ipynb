```python
import numpy as np
import matplotlib.pyplot as plt
from Higgs_R2.background_time import Background
from Higgs_R2.Potential import Potential_function
plt.style.use('style')
```

```python
from pathlib import Path
fig_dir = Path.cwd().parents[1] / "Paper" / "Figures"
fig_dir.mkdir(parents=True, exist_ok=True)

def savefig(name, ext="pdf"):
    from pathlib import Path
    import matplotlib.pyplot as plt

    fig_dir = Path.cwd().parents[1] / "Paper" / "Figures"
    fig_dir.mkdir(parents=True, exist_ok=True)
    path = fig_dir / f"{name}.{ext}"
    plt.savefig(path, bbox_inches="tight", dpi=300, pad_inches = 0.1)
    print(f"âœ… Figura guardada en {path}")
```

```python
V = 'exp(-sqrt(8/3)*phi)*(lambda_h*h**4/4 + (1/(16*xi_s))*(exp(sqrt(2/3)*phi) - 1 - xi*h**2)**2)'
parameters = {'lambda_h': 1e-10, 'xi': 1e-1 , 'xi_s': 4e8}
potential = Potential_function.from_string(V, param_values=parameters)
```

```python
bg = Background(potential, phi0= 5.7, h0 = 1e-4, lambda_h= 1e-10, xi = 1e-1, xi_s = 4e8)
```

```python
vars = ['t', 'N', 'phi', 'y', 'h', 'z', 'H', 'a', 'aH', 'eps_H', 'dot_sigma', 'eta_perp', 'eta_sigma', 'm_iso']
t, N, phi, y, h, z, H, a, aH, eps_H, dot_sigma, eta_perp, eta_sigma, m_iso = (bg.data[i] for i in vars)
Nend = bg.N_end
Nend
```

```python
plt.semilogy(N, abs(eta_sigma), label = r"$|\eta_{||}|$")
plt.semilogy(N, abs(eta_perp), label = r"$|\eta_\perp|$")

plt.axvline(25, linestyle = 'dashed', linewidth= 1, color = 'k', alpha  = 0.6)
plt.axvline(38, linestyle = 'dashed', linewidth= 1, color = 'k', alpha  = 0.6)

plt.xlabel(r"$N_e$")

plt.xlim(2, Nend)
plt.ylim(1e-6, 1e1)
plt.legend()
Ne_turn_ini = 25
Ne_turn_end = 38

plt.axvspan(
    Ne_turn_ini,
    Ne_turn_end,
    color = 'gray',
    alpha=0.15,
    zorder=0,
    label="Turning phase"
)
#savefig('eta1')

```

```python
plt.semilogy(N, eps_H)
plt.axvline(25, linestyle = 'dashed', linewidth= 1, color = 'k', alpha  = 0.6)
plt.axvline(38, linestyle = 'dashed', linewidth= 1, color = 'k', alpha  = 0.6)

plt.xlabel(r"$N_e$")
plt.ylabel(r"$\epsilon_H$")
plt.xlim(2, Nend)
plt.ylim(1e-6, 1e0)

Ne_turn_ini = 25
Ne_turn_end = 38

plt.axvspan(
    Ne_turn_ini,
    Ne_turn_end,
    color = 'gray',
    alpha=0.15,
    zorder=0,
    label="Turning phase"
)

savefig('eps_H1')


```
